```python
stopw=stopwords.words("english")
lem=WordNetLemmatizer()
```

```python
#creating a function  to lowercase the text,remove stopswords and lemmatizer 
def preprocess(text):
    x=text.lower()
    l=[]
    for word in word_tokenize(x):
        if word in stopw or len(word)<3:
            pass
        else:
            word=lem.lemmatize(word,pos="v")
            l.append(word)
    return " ".join(l)
```

```python
preprocess(df["file_content"][0])
```

```python
###applying on all the file_content column
df["file_content"]=df["file_content"].progress_apply(preprocess)
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(df.file_content)
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Concatenate the original DataFrame with the TF-IDF DataFrame
df = pd.concat([df, tfidf_df], axis=1)
```

```python
print(df.shape)
df.head()
```

```python
def find_similar_documents(query_text, documents_df, vectorizer, tfidf_matrix):
    query_vector = vectorizer.transform([query_text])
    cosine_similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()
    similar_document_indices = cosine_similarities.argsort()[::-1]
    similar_documents = documents_df.iloc[similar_document_indices]
    return similar_documents
```

# User Query - 1

```python
# tfidf_vectorizer and tfidf_matrix are already created
query = input('Search : ')
similar_records = find_similar_documents(query, df, tfidf_vectorizer, tfidf_matrix)

# Display the top N similar records
top_n = int(input('Enter the No.of Search Results you want :'))
top_similar_records = similar_records.head(top_n)
print(f"Top {top_n} similar records for query '{query}':")
print(top_similar_records[['name', 'similarity']])
```

# Applying chunking to improve the performance

```python
# Function to split paragraph into chunks with overlapping text
def split_paragraph(paragraph, chunk_size=500, overlap=20):
    chunks = []
    words = paragraph.split()
    start_idx = 0
    while start_idx < len(words):
        end_idx = min(start_idx + chunk_size, len(words))
        chunk = " ".join(words[start_idx:end_idx])
        chunks.append(chunk)
        start_idx += chunk_size - overlap
    return chunks

# Apply split_paragraph function to each paragraph
df['chunks'] = df['file_content'].progress_apply(lambda x: split_paragraph(x)) 

df["chunks"][0]
```

# Initializing lists to store extracted data

```python
# Initialize lists to store extracted data
documents = []
metadatas = []
ids = []

for index, row in df.iterrows():
    # Extract data from the 'name' column (assuming 'name' is the column of interest)
    document_name = row['name']
    item_id = row['num']  # Assuming 'num' contains item_id or metadata

    # Append extracted data to respective lists
    documents.append(document_name)
    metadatas.append({"item_id": item_id})
    ids.append(str(index + 1))  # Using 1-based index as ids

# Print first few entries from each list for verification
print("Documents:", documents[:5])
print("Metadatas:", metadatas[:5])
print("IDs:", ids[:5])
```

# Storing Embeddings into ChromaDB

```python
# client = chromadb.Client()
import chromadb
from chromadb.utils import embedding_functions

chroma_client = chromadb.PersistentClient(path="my_chromadb")
```

```python
sentence_transformer = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")

collection = chroma_client.get_or_create_collection(name="my_collection", 
                                                        embedding_function=sentence_transformer,
                                                        metadata ={"hnsw:space":"cosine"})
```

# NOTE : The data we're trying to add to the collection exceeds the maximum batch size allowed by the chromadb library (specifically, a maximum batch size of 166 in this case). 
# To resolve this issue, we need to split our data into smaller batches that fit within the allowed size limit.

```python
# Name of the collection
collection
```

```python
# Define maximum batch size (adjust as per your needs)
max_batch_size = 166

# Split data into batches and add to collection
for i in range(0, len(documents), max_batch_size):
    batch_documents = documents[i:i+max_batch_size]
    batch_metadatas = metadatas[i:i+max_batch_size]
    batch_ids = ids[i:i+max_batch_size]

    # Add batch to the collection
    collection.add(documents=batch_documents, metadatas=batch_metadatas, ids=batch_ids)
    print(f"Batch {i//max_batch_size + 1} added to collection.")
    print("Batch of Data added to collection successfully.")

print("All data added to collection successfully.")
```
